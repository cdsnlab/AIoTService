```python
hh_folders=sorted(glob.glob("./dataset/hh/*"))
print(len(hh_folders))

dict_={}

for fn, folder in enumerate(hh_folders):
    data_name=folder.split(".")[1][-5:]
    with open("{}/ann.txt".format(folder),'rb') as f:
        text=f.readlines()
    print(folder, data_name)

    data=np.array(read_hh(text))
    data=np.array([datum for datum in data if datum[0][0] in ['M','D']])
    data_mdn=[]
    for datum in data:
        if datum[0][0]=='M' and datum[1]=='OFF':
            continue
        data_mdn.append(datum)
    data=np.array(data_mdn)

    filedir="./preprocessed/hh/{}".format(data_name)
    if not os.path.exists(filedir):
        os.mkdir(filedir)

    np.save("{}/pp.npy".format(filedir), data)

    sensors=sorted(set(data[:,0]))
    activities=sorted(set(data[:,-1]))
    print("""
        {} Sensors: {}
        {} Activities: {}
    """.format(len(sensors), sensors, len(activities), activities))
```
